```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
file = "space_missions.csv"
data = pd.read_csv(file, encoding='ISO-8859-1')
```

```python
data.head()
```

```python
df = data
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['Date']= pd.to_datetime(df['Date'])
```

```python
df1 = df.drop(['Time', 'Price'], axis=1 )
```

```python
df1[['Location','Country']] = df1.Location.apply(
   lambda x: pd.Series(str(x).rsplit(",", 1)))
```

```python
df1.head()
```

```python
df1["Country"].value_counts(normalize=True)
```

```python
df1['Country'] =df1['Country'].str.replace('Kazakhstan','Russia')
df1['Country'] =df1['Country'].str.replace('New Mexico','USA')
df1['Country'] =df1['Country'].str.replace('Pacific Ocean','USA')
df1['Country'] =df1['Country'].str.replace('Barents Sea','Russia')
df1['Country'] =df1['Country'].str.replace('Yellow Sea','China')
df1['Country'] =df1['Country'].str.replace('Gran Canaria','USA')
df1['Country'] =df1['Country'].str.replace('Shahrud Missile Test Site','Iran')
df1['Country'] =df1['Country'].str.replace('Pacific Missile Range Facility','USA')
df1
```
